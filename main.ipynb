## Math Function Graphics

```python
# Imports
import matplotlib.pyplot as plt
```

```python
def plot_creator(function,x_range):
    fig, ax = plt.subplots()
    ax.axis("off")
    y_axis = [function(val)*1000 for val in x]
    y3 = [0 for _ in x]
    y4 = [i for i in range(int(min(y_axis))-500,int(max(y_axis))+500)]
    zero_x = [0 for _ in range(int(min(y_axis))-500,int(max(y_axis))+500)]
    plt.plot(x,y3,color="black")
    plt.plot(zero_x,y4,color="black")
    
    
    plt.plot(x_range,y_axis)
    return plt
```

```python
# y = kx
MAX_X = 50
MIN_X = -50
x = list(range(MIN_X,MAX_X))

def linear(k=1):
    def func_body(x):
        return k*x
    return func_body

lin_plt = plot_creator(linear(k=1),x)
lin_plt.show()
```

```python
# y = kx^2+b
MAX_X = 10
MIN_X = -10
x = list(range(MIN_X,MAX_X))

def quadratic(k=1,b=0):
    def func_body(x):
        return k*(x**2)+b
    return func_body

quad_plot = plot_creator(quadratic(k=10,b=-100),x)
quad_plot.show()
```

```python
# y = sin(x)
from math import pi,sin
MAX_X = 360
MIN_X = -360
x = list(range(MIN_X,MAX_X))

def sinusoid(x):
    return sin(x*pi/180)


sin_plot = plot_creator(sinusoid,x)
sin_plot.show()
```

```python
# y = ln(x)
from math import log
MAX_X = 200
MIN_X = 1
x = list(range(MIN_X,MAX_X))

def logarithmic(x):
    return log(x)

log_plot = plot_creator(logarithmic,x)
log_plot.show()
```
